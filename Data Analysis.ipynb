# Average Game Completion

```python
import game
from dm.simulator import Simulator
from dm.dm_random import Random
from dm.dm_closest import Closest
from dm.dm_mcts import MCTS
from dm.dm_belief_sampling import BeliefSampling

from copy import deepcopy
```

```python
randos = game.create(count=85)
repeatSim = 3
```

```python

p = 15
for rando_idx in range(len(randos)):
    rando = randos[rando_idx]        
    try:
        for i in range(repeatSim):
            sim_rand = Simulator(deepcopy(rando),Random)
            sim_close = Simulator(deepcopy(rando),Closest)
            sim_mcts = Simulator(deepcopy(rando),MCTS)
            sim_bsamp = Simulator(deepcopy(rando),BeliefSampling)

            sim_mcts.agent.decisionmaker.numParticles = 50
            sim_mcts.agent.decisionmaker.setSimDepth = (5, 3)
            sim_mcts.agent.decisionmaker.m = 5

            sim_rand.simulate(outputFile=f'data/random/rand_p{p}_i{i}.txt')
            sim_close.simulate(outputFile=f'data/closest/close_p{p}_i{i}.txt')
            sim_mcts.simulate(outputFile=f'data/mcts/mcts_p{p}_i{i}.txt')
            sim_bsamp.simulate(outputFile=f'data/belief_sampling/bsamp_p{p}_i{i}.txt')
            
    except ValueError:
        randos = game.create(count=(100 - p))
        continue
        
    p += 1
```

```python
import os

rand_times = []

for filename in os.listdir("data/random"):
    if filename.endswith(".txt"): 
        with open(os.path.join("data/random", filename), 'r') as f:
            random_time = f.readline();
            i = random_time.find(':')
            random_time = random_time[i+2:]
            rand_times.append(float(random_time))

```

```python
close_times = []

for filename in os.listdir("data/closest"):
    if filename.endswith(".txt"): 
        with open(os.path.join("data/closest", filename), 'r') as f:
            closest_time = f.readline();
            i = closest_time.find(':')
            closest_time = closest_time[i+2:]
            close_times.append(float(closest_time))
```

```python
mcts_times = []

for filename in os.listdir("data/mcts"):
    if filename.endswith(".txt"): 
        with open(os.path.join("data/mcts", filename), 'r') as f:
            mcts_time = f.readline();
            i = mcts_time.find(':')
            mcts_time = mcts_time[i+2:]
            mcts_times.append(float(mcts_time))
```

```python
bsamp_times = []

for filename in os.listdir("data/belief_sampling"):
    if filename.endswith(".txt"): 
        with open(os.path.join("data/belief_sampling", filename), 'r') as f:
            bsampling_time = f.readline();
            i = bsampling_time.find(':')
            bsampling_time = bsampling_time[i+2:]
            bsamp_times.append(float(bsampling_time))
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
rand_y = np.ones(len(rand_times))
close_y = 2*np.ones(len(close_times))
mcts_y = 3*np.ones(len(mcts_times))
bsamp_y = 4*np.ones(len(bsamp_times))

y = np.concatenate((rand_y, close_y))
y = np.concatenate((y, mcts_y))
y = np.concatenate((y, bsamp_y))

x = rand_times + close_times + mcts_times + bsamp_times

rand_avg = np.mean(rand_times)
close_avg = np.mean(close_times)
mcts_avg = np.mean(mcts_times)
bsamp_avg = np.mean(bsamp_times)

ys = [1, 2, 3, 4]
labels = ['random', 'closest', 'mcts', 'belief sampling']

plt.scatter(x,y)
plt.yticks(ys, labels)

print(rand_avg)
print(close_avg)
print(mcts_avg)
print(bsamp_avg)
```
